```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv("supermarket_short.csv")

def process_data(data_param, min_supp, min_conf):
    frequent_itemsets = apriori(data_param, min_support=min_supp, use_colnames=True, low_memory=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
    rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
    rules = rules.drop(["lift", "leverage", "conviction", "antecedent support", "consequent support"], axis=1)
    rules = rules[rules['confidence'] >= min_conf]
    return rules
```
